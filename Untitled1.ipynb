```python
from guitar_scraping.questions import articles_mean_price
import pandas as pd
df = articles_mean_price()
df.head()
```

```python

display(df.columns.values)
relevant_cols = ['avg_price', 'bauweise', 'cutaway', 'farbe',
       'griffbrett', 'inkl_gigbag', 'koffer', 
       'hersteller_id', 'buende',
       'holz_decke', 'holz_boden_zargen']


wood_cols = ['holz_decke', 'holz_boden_zargen','griffbrett']

cat_cols = ['hersteller_id','bauweise','farbe']+wood_cols
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
woods = df[wood_cols].values.flatten()
label_encoder.fit(woods)
build_encoder = LabelEncoder()

display(max(label_encoder.transform(woods)))
display(len(set(woods)))
display(len(woods))

data = df[relevant_cols].copy()
data.bauweise = build_encoder.fit_transform(data.bauweise.fillna('Unknown'))
data.farbe = data.farbe.astype('category')
data.buende.fillna(int(data.buende.median()),inplace=True)
for c in wood_cols:
    data[c] = label_encoder.transform(data[c])
data.bauweise = build_encoder.fit_transform(data.bauweise)
data.info()
```

```python
y = data.avg_price
X = data[relevant_cols[1:]]
X = pd.get_dummies(X,columns = cat_cols)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

Xtrain, Xtest = train_test_split(X,test_size=.1)
ytrain, ytest = train_test_split(y,test_size=.1)

r = RandomForestRegressor()

p = Pipeline([
('scaler',StandardScaler()),
('regressor', r)
])

r.fit(Xtrain, ytrain)

pred = r.predict(Xtest)
for p,t in zip(pred,ytest):
    print(p,t)
    
```
