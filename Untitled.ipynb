```python

import pandas as pd
data = pd.read_csv('data/2019-12-25/folk_gitarren_chunk_000.csv').preis.values
dist = s.rayleigh
params = dist.fit(data)
print(params)


x = np.linspace(params[0]-5*params[1],params[0]+5*params[1],400)
plt.hist(data,bins=20,density=True)
plt.plot(x,dist.pdf(x,*params))
plt.show()
```

```python
foo = lambda x,a,b : b*x+2*a 

data2 = np.linspace(0,500,1000)+np.random.uniform(-10,10,1000)+20
data = data2

from scipy.optimize import curve_fit
params = curve_fit(foo,range(len(data)),data)

print(params[0])
print(np.sqrt(np.abs(params[1])).diagonal())
plt.scatter(np.arange(len(data)),data)
x = np.linspace(0,len(data),200)
plt.plot(x,foo(x,*params[0]))
```
